```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
url = 'https://www.forbes.com/sites/christopherelliott/2019/10/16/20-best-places-visit-in-2020/#7310d97f6251'
html = requests.get(url).content
html
```

```python
soup = BeautifulSoup(html, 'lxml')
soup.prettify()
```

```python
cities = soup.find_all('strong')
cities
```

```python
num = list(range(1,21))
rank = [city.text.strip() for city in cities]
rank
```

```python
cities_clean = [item.replace('\n','').replace(' ','').replace('Why?','').replace('Dont miss','').replace('Who should go?','') for item in rank]
```
