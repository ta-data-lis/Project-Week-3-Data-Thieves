```python
#url = "https://supercasa.pt/comprar-casas/lisboa"
url = "https://supercasa.pt/comprar-casas/lisboa/parque-das-nacoes/"
response = requests.get(url)
response
```

```python
response.content
```

```python
soup = BeautifulSoup(response.content)
```

```python
homes = soup.find_all("div", attrs ={"class":"property big-picture"})
homes[0].get_text().strip()
```

```python
price = soup.find_all("div", attrs ={"class":"property-price"})
price[0].get_text().strip()
```

```python
size = soup.find_all("div", attrs ={"class":"property-features"})
lista = size[0].get_text().strip().split()
lista
```

```python
new_list[]
for x in lista:
    if x.isnumeric():
        new_list.append(x)
    else:
        pass
new_list       
```

```python
zone = soup.find_all("h2", attrs ={"class":"property-list-title"})
zone[3].get_text().strip()
```

```python
lisbon = {}
i = 0
for i in homes:
    price = i.soup.find_all("div", attrs ={"class":"property-price"})
    size = i.find_all("div", attrs ={"class":"property-features"})
    zone = i.find_all("h2", attrs ={"class":"property-list-title"})
    lisbon[i] = {"Price":price,
                      "Size": size,
                      "Zone": zone}
    i+=1
lisbon    
```
