```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# tag: table
# class: wikitable sortable jquery-tablesorter
url = 'https://en.wikipedia.org/wiki/List_of_Academy_Award-winning_films'
html = requests.get(url).content
soup = BeautifulSoup(html, "lxml")

#table_body = [element.text for element in soup.find_all('table')]
table_body = soup.find_all('tbody')[0]
rows = [element.text for element in table_body.find_all('tr')]

clean_rows = [row.strip().replace('\n','  ').split('  ') for row in rows]

colnames = ['Title', 'Year', 'Awards', 'Nominations']
oscars = pd.DataFrame(clean_rows[1:], columns = colnames)
oscars
```

```python
test = oscars

#test['Year'] = test['Year'].astype(str)
#test['Year'] = test['Year'].split("/")

#test = test['Year'].astype(int)



# Like this it matches the movies' file year range
test = test[(test['Year'] <= '2016') & (test['Year'] >= '1986')]


# Resetting the index, as it was starting at 31
test.reset_index(drop = True, inplace = True)

test.head()
```
