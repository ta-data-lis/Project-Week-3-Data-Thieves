```python
import ssl
import urllib.request
import json
import codecs
import csv

# Put your TMDB API key as text string here
TMDB_API_KEY = "5d6ee3c337c9ed1dd8636a0112113fd2"

# The Avengers: Endgame movie
TMDB_MOVIE_ID = 299534

# Create context for HTTPS connection
context = ssl._create_unverified_context()

# Open URL request for downloading 1st page of reviews (first 20 reviews afaik)
u = urllib.request.urlopen("https://api.themoviedb.org/3/movie/%i/reviews?api_key=%s&language=en-US&page=1" % (TMDB_MOVIE_ID, TMDB_API_KEY), context=context)

# Read JSON file with results from TMDB
raw_data = u.read()

# Convert JSON file to python dict
data = json.loads(raw_data)

# Declare CSV field names for export (same as dict keys of TMDB reviews returned by API)
fieldnames = ["author","content","id","url"]

# Open output CSV file with UTF-8 encoding, filename is TMDB ID of the movie with ".csv" extension --> "299534.csv"
f = codecs.open("%i.csv" % TMDB_MOVIE_ID, "w", "utf-8-sig")

# Initiate CSV writer with tabs as field delimiters
writer = csv.DictWriter(f, dialect="excel", delimiter="\t", fieldnames=fieldnames)

# Write CSV header (field names)
writer.writeheader()

# Go through all reviews
for review in data["results"]:

    # Prepare empty row
    row = {}

    # Go through all keys of review
    for key in fieldnames:

        # For "content" field we need to remove line breaks (\n\r) and tabs (\t), which would mess CSV
        if key == "content":

            # Put text string with review content without line breaks and tabs into CSV field
            row[key] = review[key].replace('\n', ' ').replace('\r', '').replace('\t', ' ')

        # All other fields will be used without conversion
        else:
            row[key] = review[key]

    # Write row with review to CSV file
    writer.writerow(row)

# Close output CSV file
f.close()
```
